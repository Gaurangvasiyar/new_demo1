```python
new_df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/hiring.csv')
new_df
```

```python
string_to_number = {
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10,
    'eleven':11
}
```

```python
new_df['experience'] = new_df['experience'].replace(string_to_number)
new_df.experience
```

```python
new_df['test_score(out of 10)']=new_df['test_score(out of 10)'].fillna(new_df['test_score(out of 10)'].median())
new_df
```

```python
new_df['experience']=new_df['experience'].fillna(new_df['experience'].median())
new_df
```

```python
new_reg=linear_model.LinearRegression()
new_reg.fit(new_df.drop('salary($)',axis='columns'),new_df['salary($)'])
```

```python
new_reg.predict([[2, 9, 6]])
```

```python
new_reg.predict([[12, 10, 10]])
```
